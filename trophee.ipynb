```python
import numpy as np
from PIL import Image
#list a image RGB
pixel_list = [
    [[255, 0, 0], [0, 255, 0], [0, 0, 255]],
    [[0, 0, 0], [255, 255, 255], [126, 126, 126]]
]
# ou on peut faire
binary_pixel_list = [
    [[0b11111111, 0b0, 0b0], [0b0, 0b11111111, 0b0], [0b0, 0b0, 0b11111111]],
    [[0b0, 0b0, 0b0], [0b11111111, 0b11111111, 0b11111111], [0b1111110, 0b1111110, 0b1111110]]
]
np_array = np.array(pixel_list, dtype=np.uint8) #convert to numpy array
image = Image.fromarray(np_array) #use numpy to create image
image.save('listimage.png') #save
```
